```python
from dotenv import load_dotenv
from src.load_pdf import load_pdf
from src.embed_store import create_vector_store
from src.retriever import retrieve_docs
from src.rag_pipeline import format_context
from langchain_openai import ChatOpenAI

load_dotenv()
pdf_path = 'data/Famous old receipts - bread.pdf'
documents = load_pdf(pdf_path)
vector_store, chunks = create_vector_store(documents)
print(f'Loaded {len(documents)} page documents and created {len(chunks)} chunks.')
```

```python
query = 'a bread that uses wheat flour and is suitable for a dinner party'
context_docs = retrieve_docs(vector_store, query, k=5)
context_text = format_context(context_docs)
system_message = f'''
You are a helpful document assistant.
Answer only from the provided context.

Context:
{context_text}
'''
llm = ChatOpenAI(temperature=0)
response = llm.invoke([('system', system_message), ('human', query)])
response.content
```
